### e) 

This boxplot shows two columns "Outstate" and "Private". Where the values of the column "Outstate" is represented on the Y-axis as cost and the values of the column "Private" is represented on the X-axis as Yes or No. 

What each box plot tells is that Private has a top quartile (Q1) that represents 25% and a bottom (Q2) that represents 25% and range in between a interquartile range (IQR) with a median as a solid line and whiskers as the outliners.   

```python
college.boxplot(column = "Outstate", by = "Private") # use boxplot method on college df on columns: "Outstate" (The out of state tuition) vs Private (indicator).

```

### f)

```python
college["Elite"] = pd.cut(college["Top10perc"],bins = [0,50,100], labels = ["No", "Yes"]) # create a new category named "Elite" based on teh column "Top10perc" and change bin range to acomandate the value of said column and categorize "Elite" as "No" or "Yes"


college["Elite"].value_counts() # count the number of "Yes2 or "No" in "ELite"-column 


college.boxplot(column = "Outstate", by = "Elite") # Creating a Box Plot for Out-of-State Tuition by Elite 
```

```python

import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8))      # Divide plot window into four regions.

plt.subplots_adjust(hspace=0.3, wspace=0.3)         # Adjusting the horizontal and vertical space between the axes.




# For books 2340 is max and 96 is min. 2340 - 96 = 2244. 2244/20 = 112.2. So there is 20 groups where each show in 112.2 intervals. 0 to (incl) 112.2, 112.3 to (incl) 224.4 and so on.
college["Grad.Rate"].plot.hist(ax=axes[0, 0], xlabel="Graduation Rate", bins=50, color="blue", title="Graduated")       # Column Grad.Rate = estimated rate of graduated students, Apps = number of applications recieved, Accept = numher applications accepted, Enroll = number of studens enrolled. 
college["Apps"].plot.hist(ax=axes[0, 1], xlabel="Applications recieved", bins=50, color="green", title="Applications")      # Example bins are starting from min to max values which are splitet in groups. Books are splitet in 20 groups and frequency shows how many fit to each group.
college["Accept"].plot.hist(ax=axes[1, 0], xlabel="Applications accepted", bins=25, color="red", title="Accepted")
college["Enroll"].plot.hist(ax=axes[1, 1], xlabel="Enrolled students", bins=60, color="orange", title="Enrolled")
```
